```python
import pandas as pd
import numpy as np
```

to read from a csv file, we use pd.read_csv()

```python
heart_db = pd.read_csv('heart.csv')
heart_db
```

to see the first n number of rows, we use .head(n)

```python
heart_db.head(3)
```

to see the last n number of rows, we use .tail(n)

```python
heart_db.tail(3)
```

.isnull() checks if each data node whether is null or not

```python
heart_db.isnull()
```

```python
heart_db.notnull()
```

but this is impractical
we can use .sum() to see how many NaNs are in each column

```python
heart_db.isnull().sum()
```

to see the columns, we use .columns

```python
heart_db.columns
```

to see the datatype of our db, we use .info()

```python
heart_db.info()
```

to see information from our data such as count mean std min or percentile, we use .describe()
.std means standard deviation. It tells you how spread out the numbers are around the mean.

```python
heart_db.describe()
```

we can also have it done to a single column

```python
heart_db['trestbps'].describe()
```

or you can do it individually

```python
heart_db['age'].mean()
```

```python
heart_db['age'].count()
```

```python
heart_db['age'].std()
```

shows the type of each column in the database

```python
heart_db.dtypes
```

we can select only the wanted columns

```python
heart_db[['age', 'sex']]
```

we can put conditions on each column

```python
heart_db[heart_db['age'] < 45]
```

we can do both simultaneously

```python
heart_db[heart_db['age'] < 45][['age', 'trestbps']]
```

we can use np.sqrt to make a new column from the square root of another column

```python
heart_db['square-1'] = np.sqrt(heart_db['age'])
```

we can use .apply(lambda x:f(x)) to do simple mathematic operations

```python
heart_db['square-2'] = heart_db['age'].apply(lambda x:x ** 2)
```

or we can intuitively just do operations to two whole columns and have them be a new column

```python
heart_db['calculation'] = heart_db['age'] ** heart_db['trestbps']
```

these two essentially do the same thing

```python
heart_db['age'] += 1
```

```python
heart_db['age'] = heart_db['age'].apply(lambda x:x + 1)
```

```python
heart_db
```

you can save the db to a csv file with .to_csv()
but doing so will also add the enumerations to the ongoing list, resulting in an extra unnamed: index column.

```python
heart_db.to_csv("heart_saved.csv")
heart_saved = pd.read_csv('heart_saved.csv')
heart_saved
```

to remedy this, we also use the kwargs index=False

```python
heart_db.to_csv("heart_saved.csv", index=False)
heart_saved = pd.read_csv('heart_saved.csv')
heart_saved
```

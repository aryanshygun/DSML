```python
import pandas as pd
```

```python
df = pd.read_csv('REPO - zero-to-mastery-ml/data/car-sales-extended.csv')
df
```

```python
df['Make'].value_counts().plot(kind='bar')
```

```python
df.groupby('Make')['Price'].mean().plot(kind='bar')
```

```python
df['Colour'].value_counts().plot(kind='bar')
```

```python
df.plot.scatter(x='Odometer (KM)', y='Price')
```

```python
df['price_per_km'] = df['Price'] / df['Odometer (KM)']
df.groupby('Make')['price_per_km'].mean().sort_values().plot(kind='bar')
```

```python
pd.crosstab(df['Make'], df['Colour']).plot(kind='bar', stacked=True)
```

```python
import seaborn as sns
sns.boxplot(x='Make', y='Price', data=df)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
```

we can use np.sqrt to make a new column from the square root of another column

```python
heart_db['square-1'] = np.sqrt(heart_db['age'])
```

we can use .apply(lambda x:f(x)) to do simple mathematic operations

```python
heart_db['square-2'] = heart_db['age'].apply(lambda x:x ** 2)
```

or we can intuitively just do operations to two whole columns and have them be a new column

```python
heart_db['calculation'] = heart_db['age'] ** heart_db['trestbps']
```

these two essentially do the same thing

```python
heart_db['age'] += 1
heart_db['age'] = heart_db['age'].apply(lambda x:x + 1)
heart_db
```

you can save the db to a csv file with .to_csv()
but doing so will also add the enumerations to the ongoing list, resulting in an extra unnamed: index column.

```python
heart_db.to_csv("heart_saved.csv")
heart_saved = pd.read_csv('heart_saved.csv')
heart_saved
```

to remedy this, we also use the kwargs index=False

```python
heart_db.to_csv("heart_saved.csv", index=False)
heart_saved = pd.read_csv('heart_saved.csv')
heart_saved
```

if we have NaN amounts in our data, we can either omit the entire row, or add placeholder numbers instaed of the NaNs

```python
broken_health_db = pd.read_csv('broken_heart.csv')
```

important, always use .sum() to see the count of nulls in a column. .count() will count all rows regardless of whether its str int or null

```python
broken_health_db.isnull().sum()
```

if we want to change the data of the NaNs in a column and have it replaced, we have two methods:

we either do assignment with .fillna('value')

```python
broken_health_db['trestbps'] = broken_health_db['trestbps'].fillna(broken_health_db['trestbps'].mean())
```

or we do it directly but it will be deprecated in future and even tho it works, it also gives an error

```python
broken_health_db['chol'].fillna(broken_health_db['chol'].mean(), inplace=True)
```

```python
broken_health_db.isnull().sum()
```

to drop a column or a row, we use .drop('row/column', axis='0/1') to delete columns we use axis = 1 and to delete rows we use axis = 0

```python
heart_db.drop(['age'], axis=1)
```

if we want to drop some of the rows based on a condition( for example the ones that contain null)

we can filter the table by only having datas in 'chol' column that their isnull is FALSE

```python
broken_health_db = broken_health_db[broken_health_db['chol'].isnull() == False]
```

or we can drop any rows with any NaNs

```python
broken_health_db.dropna(inplace=True)
```

or we can have specific columns checked out

```python
broken_health_db.dropna(subset=['chol', 'age'], inplace=True)
```

```python
broken_health_db
```

after deleting rows, we use reset_index() in order to have the rows index numbers cleaned out

```python
broken_health_db.reset_index(drop=True, inplace=True)
```

```python
broken_health_db
```

we use .sample to sample a part of the data in order to feed it into the model. it is detrimental because it shuffles the data and stops any sort of bias in the model incase the data is sorted by a specific feature.

```python
broken_health_db.sample(frac=1)
```

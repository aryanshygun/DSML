we use numpy because its faster, and well optmized in comparison with raw python

NumPy arrays are implemented in C, meaning operations happen at compiled-speed, not interpreted Python-speed.

A loop over a list in Python is painfully slow compared to a vectorized operation in NumPy.

NumPy can be 10–100x faster than native Python lists for numeric tasks.

```python
xlist = np.random.randn(10000,3)
```

we can use the built in python function sum() to add the values

```python
sum(xlist)
```

```python
np.sum(xlist)
```

we can use the axis=0 arguement to make it so that it doesnt add all the inner indexes, but respects the first indexes

```python
np.sum(xlist, axis=0)
```

we can use %timeit to time the duration it took to run that function

```python
%timeit sum(xlist)
%timeit np.sum(xlist, axis=0)
```

as it can be seen, using the np.sum() function results in a much faster time period for the action.

```python
%timeit -n 100 sum(xlist)
%timeit -n 200 np.sum(xlist)
```

at the end of each result of %timeit, it says 100 loops each or 100000 loops each. thats not about the function and its about the timeit magic function. it is essentially running the mentioned function a number of times to get the average of each procedure. we can use %timeit -n x, to force the magic function to only run x amount of time.

```python
high_var_array = np.array([-100, -50, 0, 50, 100])
low_var_array = np.array([-4, -2, 0, 2, 4])
```

```python
np.var(high_var_array), np.var(low_var_array)
```

```python
np.std(high_var_array), np.std(low_var_array)
```

as we can see, both of the arrays have the mean of 0, but one is much further apart meanwile two are much closer to each other. when we want to calculate their variance or standard deviation, we see the difference

the closer the standard deviation is to 0, the closer the numbers are to each other and vice versa, the higher the std is, the more spread the indexes of the list are.

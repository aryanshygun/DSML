```python
import pandas as pd
import matplotlib.pyplot as plt
```

In pandas, we use Series({}) to create lists, it enumerates the list and gives its dtype

```python
time = pd.Series([0,1,2,3,4,5,6,7])
l1 = pd.Series([4,5,7,5,6,8,3,5])
l2 = pd.Series([3,4,6,3,9,4,6,7])
l1
```

here we see how to use the pd.DataFrame({}) to add multiple Series to create a 2D plot.

This uses both matplotlib and pandas

```python
data = pd.DataFrame({"l 1":l1, "l2":l2, "time":time})
data
```

we can access a single column of the DataFrame by calling it data.column_name.

In this case we can write data.l2 or data.time

if the name has whitespace, we can use the traditional calling method of dictionaries such as data['l 1'] or data['time']

```python
data['l 1']
```

```python
data.l2
```

we can also add columns to the data with this way

```python
data['distance'] = 0
data
```

we can also do calculations with the rows

```python
data['percentage'] = (data.l2 / data.time) / 100
data
```

we use data.plot() to convert the column and rows to borders

```python
data.plot()
```

if we want to use two Series to create a border, we use the matplotlib.pyplot library and have the code
plt.plot('x axis', 'y axis') and use plt.show() to view it. 

```python
plt.plot(time,l1)
plt.show()
```

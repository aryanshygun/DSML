`np.corrcoef(x, y)` - computes the Pearson correlation coefficient between arrays x and y.

`Pearson Correlation Coefficient (r)` - Measures the linear relationship between two variables

Range: -1 to +1

+1: Perfect positive linear correlation

-1: Perfect negative linear correlation

0: No linear relationship

|x|y|Interpretation|
|-|-|-|
|[1, 2, 3, 4]|[10, 20, 30, 40]|r â‰ˆ +1 (They rise together)|
|[1, 2, 3, 4]|[40, 30, 20, 10]|r â‰ˆ -1 (One rises, the other falls)|
|[1, 2, 3, 4]|[5, 1, 8, 0]|r â‰ˆ 0 (No linear pattern)|

The `np.mean()`,  `np.var()`, and `np.std()`  functions can be used
to compute the mean, variance, and standard deviation of arrays.  These functions are also
available as methods on the arrays.

- `Average`:
    - Definition: A general term for a central value of a dataset. It can refer to mean, median, or mode (context-dependent).

- `Mean`:
    - Definition: Sum of all values divided by the number of values.
    - Example: For [1,2,3,4,5], mean=(1+2+3+4+5)/5
    - Type: A specific type of average.

- `Median`:
    - Definition: The middle value when data is sorted. For even-sized datasets, itâ€™s the average of the two middle values.
    - Example: For [1,3,5,7,9], median=5. For [1,2,3,4,5,6], median=(3+4)/2=3.5

- `Mode`:
    - Definition: The most frequently occurring value(s) in the dataset. If all values occur equally, thereâ€™s no mode.
    - Example: For [1,2,2,3,4], mode=2. For [1,2,3,4], thereâ€™s no mode.

```python
y = rng.standard_normal(10)
print('mean:', np.mean(y))
print('median:', np.median(y))
```


### Example Data:
`[1, 3, 4, 6]`

### Step 1: Calculate the Mean
(1 + 3 + 4 + 6) / 4 = 3.5

### Step 2: Calculate the Variance
1. Find the squared differences from the mean for each value:
   - ((1 - 3.5)^2 = 6.25)
   - ((3 - 3.5)^2 = 0.25)
   - ((4 - 3.5)^2 = 0.25)
   - ((6 - 3.5)^2 = 6.25)

2. Calculate the average of these squared differences (Variance):

Variance= (6.25 + 0.25 + 0.25 + 6.25) / 4 = 3.25


### Step 3: Calculate the Standard Deviation (STD)

STD = sqrt(Variance) = sqrt(3.25) -> 1.80

### Summary:
- **Variance**: \(3.25\) (the average of squared differences from the mean).
- **Standard Deviation**: \(1.80\) (the square root of variance, in the same units as the data).

This means the data points are, on average, about **1.80 units** away from the mean of **3.5**.

```python
print('var:', np.var(y))
print('std:', np.std(y))
```

This means in the y array, on average data points are 0.38 units away from the mean of -0.44

## ðŸ“Š Plotting Basics with fig and ax

- figsize=(5, 5) â†’ sets figure size in inches (width, height).

- fig.set_size_inches(12, 3) â†’ overrides the initial figsize.

- x, y = rng.standard_normal(10) â†’ generates 10 random numbers each for x and y.

- ax.plot(x, y) â†’ plots points and connects them with lines.

- ax.plot(x, y, linestyle='', marker='o') â†’ shows dots only, no lines.

- Marker styles: 'o', 'x', 'd', 's', '+', etc.

- ax.set_xlim(-3, 3) / ax.set_ylim(-3, 3) â†’ sets visible axis ranges.

- ax.set_xlabel('x') / ax.set_ylabel('y') â†’ labels the x and y axes.

- ax.set_title('My Plot') â†’ sets the title of the subplot.

- Add ; after a plot (e.g. ax.plot(...);) to suppress extra output in Jupyter.

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.default_rng(1303)
x = rng.standard_normal(30)
y = rng.standard_normal(30)

fig, ax = plt.subplots(4, 2, figsize=(8, 12))

ax[0, 0].scatter(x, y, marker='.')
ax[0, 1].scatter(x, y, marker='o')
ax[1, 0].scatter(x, y, marker='x')
ax[1, 1].scatter(x, y, marker='s')
ax[2, 0].scatter(x, y, marker='d')
ax[2, 1].scatter(x, y, marker='+')
ax[3, 0].plot(x, y, '-')
ax[3, 1].plot(x, y, '-', marker='o')

for i in ax.flat:
    i.set_xlim(-3, 3)
    i.set_ylim(-3, 3)
    i.set_xlabel('x')
    i.set_ylabel('y')

plt.tight_layout()
plt.show()
```

We now create some more sophisticated plots. The 
`ax.contour()` method  produces a  *contour plot* 
in order to represent three-dimensional data, similar to a
topographical map.  It takes three arguments:

* A vector of `x` values (the first dimension),
* A vector of `y` values (the second dimension), and
* A matrix whose elements correspond to the `z` value (the third
dimension) for each pair of `(x,y)` coordinates.

To create `x` and `y`, weâ€™ll use the command  `np.linspace(a, b, n)`, 
which returns a vector of `n` numbers starting at  `a` and  ending at `b`.

```python
fig, ax = plt.subplots(4, 2, figsize=(8, 16), sharex=True, sharey=True)
x = np.linspace(-np.pi, np.pi, 50)
y = x
f1 = np.multiply.outer(np.cos(y), 1 / (1 + x**2))
f2 = np.multiply.outer(x, np.cos(y) + 3)
f3 = np.multiply.outer(x, 1 / (1 + y**2))
f4 = np.multiply.outer(x, np.sin(y))

ax[0, 0].contour(x, y, f1, levels=10)
ax[0, 1].imshow(f1, extent=(-np.pi, np.pi, -np.pi, np.pi))

ax[1, 0].contour(x, y, f2, levels=40)
ax[1, 1].imshow(f2, extent=(-np.pi, np.pi, -np.pi, np.pi))

ax[2, 0].contour(x, y, f3)
ax[2, 1].imshow(f3, extent=(-np.pi, np.pi, -np.pi, np.pi))

ax[3, 0].contour(x, y, f4)
ax[3, 1].imshow(f4, extent=(-np.pi, np.pi, -np.pi, np.pi))

plt.tight_layout()
```

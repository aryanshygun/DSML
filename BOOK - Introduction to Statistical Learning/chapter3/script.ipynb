```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data = {
    "Hours Studied": [0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5,
                      5.5, 6, 6.5, 7, 7.5, 8, 8.5, 9, 9.5, 10],
    "Test Score": [20, 35, 40, 45, 48, 50, 55, 57, 60, 65,
                   68, 70, 73, 75, 78, 80, 82, 85, 87, 90]
}
df = pd.DataFrame(data)
df
```

```python
x = df[["Hours Studied"]]
y = df[['Test Score']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
plt.scatter(x, y, color="blue", label="Actual")
plt.plot(x, model.predict(x), color="red", label="Predicted")
# plt.xlabel("Hours Studied")
# plt.ylabel("Test Score")
# plt.title("Linear Regression: Hours Studied vs Test Score")
plt.legend()
plt.show()
```
